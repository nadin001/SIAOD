####  Бинарный поиск

```python

    

def binary_search (elem,l):
    left = -1
    right = len(l)
    while right > left + 1:
        mid = (left + right) // 2 
        if l[mid] > elem: 
            right = mid 
        else: 
            left = mid
    if left >=0 and l[left] == elem:
        return left

tik2 = time.perf_counter()
binary = binary_search(ch, l)
tok2 = time.perf_counter()

diff2 = tok2-tik2
if binary > 0:
    print(f"Вычисление заняло {tok2 - tik2:0.10f} секунд, позиция элемента: {binary}")
    
else:
    print("Число отсутствует в массиве")
```

####  Бинарное дерево

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
    def insert(self, value):
        if self.value: 
            if value < self.value:
                if self.left is None:
                    self.left = Node(value)
                else:
                    self.left.insert(value)
            elif value > self.value:
                if self.right is None:
                    self.right = Node(value)
                else:
                    self.right.insert(value)
        else:
            self.value = value
    #Сравниваем значения с узлами
    def findval(self, val):
        if val < self.value:
            if self.left is None:
                return str(val)+" not found"
            return self.left.findval(val)
        elif val > self.value:
            if self.right is None:
                return str(val) + " not found"
            return self.right.findval(val)
        else:
            return str(self.value) + " is found"
        
l = []
# l.append(randint(-100, 100))
root = Node(15)

for i in range(10):
    l.append(randint(-100, 100))
    root.insert(l[i])

print()
print(l)
print(root.findval(int(input("Введите искомое число: "))))
            
```

####  Фибоначиев поиск

```python
def fibonacci_search(l, num):
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(l)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(l)-1))
        if (l[i] < num):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (l[i] > num):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if(fibM_minus_1 and index < (len(l)-1) and l[index+1] == num):
        return index+1
    return -1
print(l)
tik3 = time.perf_counter()
fib = fibonacci_search(l, ch)
tok3 = time.perf_counter()

diff3 = tok3-tik3
if binary > 0:
    print(f"Вычисление заняло {tok3 - tik3:0.10f} секунд, позиция элемента: {fib}")
    
else:
    print("Число отсутствует в массиве")
```

#### Интерполяционный поиск

```python
def interpolation_search(l, num):
    low = 0
    high = (len(l) - 1)
    while low <= high and num >= l[low] and num <= l[high]:
        index = low + int(((float(high - low) / ( l[high] - l[low])) * ( num - l[low])))
        if l[index] == num:
            return index
        if l[index] < num:
            low = index + 1
        else:
            high = index - 1
    return -1

print(l)
tik4 = time.perf_counter()
inter = interpolation_search(l, ch)
tok4 = time.perf_counter()

diff4 = tok4-tik4
if binary > 0:
    print(f"Вычисление заняло {tok4 - tik4:0.10f} секунд, позиция элемента: {inter}")
    print(1)
else:
    print("Число отсутствует в массиве")
```

#### Рехэширование с помощью псевдослучайных чисел

```python
import random

class HashTable:
    keys = []
    hashes = []
    
    def __init__(self, arr):
        self.appendArray(arr)
        
    def appendKey(self, item, length):
        if (length == None):
            length = len(self.keys)
            
        self.checkKeyExists(item)
        temp = hash(item)%length
        
        if self.checkCollisions(temp):
            temp = self.pseudoRandomRehash(item, length)
            
        self.hashes.append(temp)
        self.keys.append(item)
        
    def printHashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()
        
    def appendArray(self, arr):
        for item in arr:
            self.appendKey(item, len(arr))
            
    def pseudoRandomRehash(self, key, length):
        possibleValues = [i for i in range(length * 2)]
        random.shuffle(possibleValues)
        while True:
            i = possibleValues.pop()
            temp = hash(key) % length + i
            if not self.checkCollisions(temp):
                return temp
    
    def checkCollisions(self, hashValue):
        return hashValue in self.hashes

    def checkKeyExists(self, key):
        if key in self.keys:
            raise Exception("Ключ '{}' уже существует".format(key))
    
keys = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
hashes = HashTable(keys)

try:
    hashes.printHashes()
except Exception as exc:
    print(exc.args)
```

#### Метод цепочек

```python
class ChainsMethod:
    hashes = {}
    values = {}
    pointer = 0
    N = 10
    
    def __init__(self, arr:dict = None):
        if arr != None:
            self.appendArray(arr)
    
    def append(self, key, value):
        keyHash = hash(key) % self.N
        if self.checkCollisions(keyHash):
            prevInd = self.hashes[keyHash]
            self.values[self.pointer] = [key, value, prevInd]
        else:
            self.values[self.pointer] = [key, value]
            
        self.hashes[keyHash] = self.pointer
        self.pointer += 1
            
    def appendArray(self, arr:dict):
        for key, value in arr.items():
            self.append(key, value)
            
    def find(self, key):
        keyHash = hash(key) % self._N
        if not keyHash in self.hashes:
            raise Exception('Item not found')
        index = self.hashes[keyHash]
        return self.recursiveFind(key, index)

    def recursiveFind(self, key, index):
        if self._values[index][0] == key:
            return self.values[index][1]
        
        if len(self._values[index]) < 3:
            raise Exception('Item not found')
            
        return self.recursiveFind(key, self.values[index][2])
            
    
    def printHashes(self):
        print("index => [key, value, previous]:")
        for key, value in self.values.items():
            print("{} => {}".format(key,value))
            
        print("\nhash => index:")
        for key, value in self.hashes.items():
            print("{} => {}".format(key,value))
        
    def checkCollisions(self, keyHash):
        return keyHash in self.hashes
    
keys = {'a':1, 'b':2, 'c':3, 'd':4}
chain = ChainsMethod(keys)
chain.printHashes()
```
